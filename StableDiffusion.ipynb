```python
import torch
import os
from diffusers import StableDiffusionPipeline

# Load Stable Diffusion
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
).to("cuda")

# Dog breeds
dog_breeds = [
    "golden retriever",
    "german shepherd",
    "labrador retriever",
    "pug",
    "husky",
    "dalmatian",
    "rottweiler"
]

# Root dataset folder
root_dir = "dog_dataset"
os.makedirs(root_dir, exist_ok=True)

images_per_breed = 5  # change to 10, 20, etc.

for breed in dog_breeds:
    # Create folder for each breed
    breed_folder = os.path.join(root_dir, breed.replace(" ", "_"))
    os.makedirs(breed_folder, exist_ok=True)

    for i in range(images_per_breed):
        prompt = f"A realistic photo of a {breed} dog"
        image = pipe(prompt, guidance_scale=7.5, num_inference_steps=30).images[0]

        image_path = os.path.join(breed_folder, f"{breed.replace(' ', '_')}_{i+1}.png")
        image.save(image_path)

        print(f"✅ Saved: {image_path}")
```
